```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
```

```python
df = pd.read_csv('NCHSData13.csv')
df.head()
```

```python
df.columns = ['Year', 'Week', '%Death P&I', 'Expected', 'Threshold', 'All Deaths', 'Pneumonia Deaths', 'Influenza Deaths']
```

```python

# df['Pneumonia%'] = df['Pneumonia Deaths']/(df['Pneumonia Deaths']+df['Influenza Deaths'])
# df['Flu%'] = df['Influenza Deaths']/(df['Pneumonia Deaths']+df['Influenza Deaths'])

# df['Overall Pneumonia%'] = df['Pneumonia Deaths']/(df['Pneumonia Deaths']+df['Influenza Deaths'])
# df['Overall Flu%'] = df['Influenza Deaths']/(df['Pneumonia Deaths']+df['Influenza Deaths'])

df['P per 100,000'] = (df['Pneumonia Deaths']/df['All Deaths'])*1e5
df['I per 100,000'] = (df['Influenza Deaths']/df['All Deaths'])*1e5


df.head()
```

```python
colors = ["blue", "light blue", "green", "light green", "purple", 'light purple', 'coral', 'black']
palette = sns.xkcd_palette(colors)
sns.palplot(palette)
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x='Week', y='All Deaths', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weelky All Cause Mortality\nUnited States: 2013-2020', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x='Week', y='%Death P&I', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weekly Death% P&I\nUnited States: 2013-2020', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x='Week', y='Pneumonia Deaths', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weekly Pneumonia Deaths\nUnited States: 2013-2020', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax= ax, x='Week', y='Influenza Deaths', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weekly Influenza Deaths\nUnited States: 2013-2020', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x='Week', y='P per 100,000', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weekly Pneumonia Deaths per 100,000\nUnited States: 2013-2020', fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x='Week', y='I per 100,000', hue='Year', data=df, legend='full', palette=palette)
plt.title('Weekly Influenza Deaths per 100,000\nUnited States: 2013-2020', fontsize=20)
plt.show()
```
